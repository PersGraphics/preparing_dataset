```python
import numpy as np
import pandas as pd

df = pd.read_csv('sample_data/WA_Fn-UseC_-Telco-Customer-Churn.csv')

# Analyze the data:
# 1)
print(len(df))
```

```python
# Convert the column ‘TotalCharges’ to a numeric format, coercing errors ifnecessary.
#df['TotalCharges'] = df['TotalCharges'].astype('float64').errors("ignore")
df['TotalCharges']=pd.to_numeric(df.TotalCharges, errors='coerce')
df.dtypes
```

```python
# Convert the column ‘churn’ to integer format (no - 0, yes - 1).
df.Churn = df.Churn.replace({"No": 0, "Yes": 1})
df['Churn'] = df['Churn'].astype('int64')
print(df.dtypes.Churn)

```

```python
# Delete the column ‘customerid’ from the dataset.
del(df['customerID'])
```

```python
# Change all column names to lowercase.
df.columns=df.columns.str.lower()
df
```

```python
# For string values change spaces to an underscore symbol.
df = df.replace(' ', '_', regex=True)
df.head()
```

```python
#Check for categorical columns how many different values they have.
df_unique = df.select_dtypes(include=['object']).nunique().to_frame().reset_index()
df_unique.columns = ['Variable','DistinctCount']

df_unique
```
